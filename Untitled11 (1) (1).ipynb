```python
cat = []
num = []
for i in df.columns:
    if df[i].dtype=='O':
        cat.append(i)
    else:
        num.append(i)
    
```

```python
for i in num:
    fig, ax=plt.subplots(1,2)
    ax[0].hist(df[i])
    ax[1].boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
for i in cat:
    vc = df[i].value_counts()
    plt.bar(vc.index, height=vc, color='c')
    plt.xlabel(i)
    plt.ylabel(i)
    plt.show()
    
```

```python
# IQR Winsorization
#Replacing upper outliers with upper fence
#Replacing lower outliers with lowper fence

for i in num:
    q1=df_resampled[i].quantile(0.25)
    q3=df_resampled[i].quantile(0.75)
    iqr=q3-q1
    lf=q1-(1.5*iqr)
    uf=q3+(1.5*iqr)
    df_resampled.loc[df_resampled[i]<lf,i]=lf
    df_resampled.loc[df_resampled[i]>uf,i]=uf
```

```python
for i in num:
    fig, ax=plt.subplots(1,2)
    ax[0].hist(df_resampled[i])
    ax[1].boxplot(df_resampled[i])
    plt.title(i)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
print("Model Performance: ")
print("Training accuracy: ",accuracy_score (y_train,y_train_pred))
print("Training precision: ", precision_score(y_train, y_train_pred))
print("Training recall: ", recall_score(y_train, y_train_pred))
print("Training confusion matrix: ",confusion_matrix(y_train, y_train_pred))
print("Training classification report: ",classification_report(y_train, y_train_pred))
```

```python
print("Model Performance: ")
print("Testing accuracy: ",accuracy_score (y_test,y_test_pred))
print("Testing precision: ", precision_score(y_test, y_test_pred))
print("Testing recall: ", recall_score(y_test, y_test_pred))
print("Testing confusion matrix: ",confusion_matrix(y_test, y_test_pred))
print("Testing classification report: ",classification_report(y_test, y_test_pred))
```
